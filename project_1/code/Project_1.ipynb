## Background

According to the [Meteorological Services Singapore](http://www.weather.gov.sg/climate-climate-of-singapore/#:~:text=Singapore%20is%20situated%20near%20the,month%2Dto%2Dmonth%20variation.), Singapore has typical tropical climate with adundant rainfall, high and uniform temperatures and high humidity all year round, since its situated near the equator. There are many factors that help us understand the climate of a country and in this project we are going to look into a few, especially rainfall.

Singapore’s climate is characterised by two main monsoon seasons separated by inter-monsoonal periods.  The **Northeast Monsoon** occurs from December to early March, and the **Southwest Monsoon** from June to September.

The major weather systems affecting Singapore that can lead to heavy rainfall are:

-Monsoon surges, or strong wind episodes in the Northeast Monsoon flow bringing about major rainfall events;

-Sumatra squalls, an organised line of thunderstorms travelling eastward across Singapore, having developed over the island of Sumatra or Straits of Malacca west of us;

-Afternoon and evening thunderstorms caused by strong surface heating and by the sea breeze circulation that develops in the afternoon.

Singapore’s climate station has been located at several different sites in the past 140 years. The station had been decommissioned at various points in the past due to changes to local land use in the site’s vicinity, and had to be relocated. Since 1984, the climate station has been located at **Changi**.

There are other metrics of climate such as temperature, humidity, sun shine duration, wind speed, cloud cover etc. All the dataset used in the project comes from [data.gov.sg](data.gov.sg), as recorded at the Changi climate station

### Coding Challenges

1. Manually calculate mean:

    Write a function that takes in values and returns the mean of the values. Create a list of numbers that you test on your function to check to make sure your function works!
    
    *Note*: Do not use any mean methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# Code:
def mean_cal(n):
    return sum(n)/len(n)

mean_cal([1,2,3,4,5,6,7,8,9])
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# Code:

def sd_cal(n):
    return ((sum([(i-mean_cal(n))**2 for i in n]))/len(n))**0.5

sd_cal([1,2,3,4,5,6,7,8,9])
```

```python
# Changing the format of wet bulb data, to include month/year and temperature only

raw_wb["wbt_date"] = pd.to_datetime(raw_wb.wbt_date)
raw_wb['month'] = raw_wb['wbt_date'].dt.strftime('%Y-%m') #Changing "month" format to YYYY-MM
wb = raw_wb.groupby("month")[["wet_bulb_temperature"]].mean() #Group by mean
wb = wb.reset_index() # Adding index
wb
```

Cleaning for raw data: (c) Number of visitor to places

```python
#  Display the data: print the first 5 rows of each dataframe to your Jupyter notebook.

raw_visit = pd.read_csv("../data/data_places.csv",header = 9)
raw_visit = raw_visit[:156] # remove remove footer at the bottom of raw file
raw_visit.columns = ["yearmonth","night_safari","river_wonders","singapore_zoo"]
raw_visit.head()
```

```python
#  Check for missing values and data types
raw_visit.info() # no missing values
```

```python
# Changing the format of "month" column to match with other datasets
# Month format from string to numeric: {"Jan":"-01", "Feb":"-02", "Mar":"-03","Apr":"-04", "May":"-05", "Jun":"-06", "Jul":"-07", "Aug":"-08", "Sep": "-09", "Oct":"-10","Nov":"-11","Dec": "-12"}

raw_visit["month"] = raw_visit["yearmonth"]
raw_visit["month"] = raw_visit["month"].str.replace(" Jan","-01")
raw_visit["month"] = raw_visit["month"].str.replace(" Feb","-02")
raw_visit["month"] = raw_visit["month"].str.replace(" Mar","-03")
raw_visit["month"] = raw_visit["month"].str.replace(" Apr","-04")
raw_visit["month"] = raw_visit["month"].str.replace(" May","-05")
raw_visit["month"] = raw_visit["month"].str.replace(" Jun","-06")
raw_visit["month"] = raw_visit["month"].str.replace(" Jul","-07")
raw_visit["month"] = raw_visit["month"].str.replace(" Aug","-08")
raw_visit["month"] = raw_visit["month"].str.replace(" Sep","-09")
raw_visit["month"] = raw_visit["month"].str.replace(" Oct","-10")
raw_visit["month"] = raw_visit["month"].str.replace(" Nov","-11")
raw_visit["month"] = raw_visit["month"].str.replace(" Dec","-12")
raw_visit["yearmonth"] = raw_visit["month"]
raw_visit = raw_visit.drop("month",axis=1)
raw_visit = raw_visit.rename(columns={"yearmonth":"month"})
```

Clearning for raw data: (d) Total monthly rainfall

```python
#  Display the data: print the first 5 rows of each dataframe to your Jupyter notebook.

raw_rainy_total = pd.read_csv("../data/rainfall-monthly-total.csv")
raw_rainy_total.head(5)
```

```python
#  Check for missing values and data types
raw_sunshine.info() # no missing values
```

Cleaning for raw data: (g) Surface Air temperature

```python
#  Display the data: print the first 5 rows of each dataframe to your Jupyter notebook.

raw_surface_air = pd.read_csv("../data/surface-air-temperature-monthly-mean.csv")
raw_surface_air.head(5)
```

```python
#  Check for missing values and data types
raw_surface_air.info() # no missing values
```

```python
# combine all the datasets to df

# merging all climate data into one dataframe
climate_data = pd.merge(wb, raw_rainy_days, on = 'month', how = 'outer') 
climate_data = pd.merge(climate_data, raw_humid, on = 'month', how = 'outer')
climate_data = pd.merge(climate_data, raw_rainy_total, on = 'month', how = 'outer')
climate_data = pd.merge(climate_data, raw_sunshine, on = 'month', how = 'outer')
climate_data = pd.merge(climate_data, raw_surface_air, on = 'month', how = 'outer')

#trim the 'month' column in raw_visit to remove spaces
raw_visit.month = raw_visit.month.str.strip() 

# combine visitor data with climate data
df = pd.merge(raw_visit, climate_data, on='month', how = 'outer') 

#There are missing values before year Mar 2013 (all visit data available from Mar 2013) and after Aug 2022 (climate data available till Aug 2022 only )
#remove missing values due to the unavailable data
df.dropna(inplace = True)
df = df.reset_index()
df.drop(columns="index",inplace=True)
df = df[df.river_wonders != "na"]
df.river_wonders = df.river_wonders.astype("float")

#creating a new column "year" in dataframe
df.month = pd.to_datetime(df.month)
df["year"] = df['month'].dt.year
df = pd.concat([df[df.columns[-1:]],df[df.columns[:-1]]],axis=1) # move the newly created column "year" to the first column
df.head(2)
```

```python
# Saved the combined dataframe to csv

df.to_csv("combined_clean_dataframe.csv")
```

Plot 3 - scatter plots

```python
places_list= [x for x in df.columns[2:5]]
climate_list= [x for x in df.columns[5:]]

fig, ax = plt.subplots(nrows=6, ncols=3, figsize=(15, 30))
col = 0
row = 0

for places in places_list:   
    for climate in climate_list:
        #df.plot(kind='scatter',x=climate, y=places,title=(places+" - "+climate))
        ax[row,col].scatter(df[climate],df[places])
        ax[row, col].set_title(places+" - "+climate)
        ax[row, col].set_xlabel(climate)
        ax[row, col].set_ylabel(places)
        row += 1
    row = 0
    col += 1
plt.tight_layout()
plt.show()
```
